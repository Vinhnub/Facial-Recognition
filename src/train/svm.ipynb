```python
import numpy as np
from sklearn.svm import SVC
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report
```

```python
data1 = np.load("E:/PythonFile/Project/Facial-Recognition/data/feature/casia_lbp_u_hsv_features.npz")
X1_train = data1["X_train"]
y1_train = data1["y_train"]
X1_test = data1["X_test"]
y1_test = data1["y_test"]

data2 = np.load("E:/PythonFile/Project/Facial-Recognition/data/feature/casia_lbp_u_rgb_features.npz")
X2_train = data2["X_train"]
y2_train = data2["y_train"]
X2_test = data2["X_test"]
y2_test = data2["y_test"]

data = np.load("E:/PythonFile/Project/Facial-Recognition/data/feature/casia_lbp_u_hsv_features.npz")

# X_train = np.concatenate((X1_train, X2_train), axis=1)
# y_train = y1_train
# X_test = np.concatenate((X1_test, X2_test), axis=1)
# y_test = y1_test

X_train = data["X_train"]
y_train = data["y_train"]
X_test = data["X_test"]
y_test = data["y_test"]

print(X_train.shape, X_test.shape)

print("Features loaded!")
print("Training SVM...")

model = Pipeline([
    ("scaler", StandardScaler()),
    ("svm", SVC(kernel="rbf", C=10, gamma="scale", probability=True, class_weight="balanced"))
])

model.fit(X_train, y_train)

print("Evaluating...")
y_score = model.predict_proba(X_test)[:, 1]
y_pred = model.predict(X_test)

print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import roc_curve
import numpy as np

fpr, tpr, thresholds = roc_curve(y_test, y_score)

fnr = 1 - tpr

eer_threshold = thresholds[np.nanargmin(np.abs(fnr - fpr))]
eer = fpr[np.nanargmin(np.abs(fnr - fpr))]

print("EER:", eer)
```
